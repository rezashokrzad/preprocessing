```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Categories and their respective fields
categories = ['Data Pipelines', 'Databases', 'Data Tools', 'Storytelling',
              'Data Visualization', 'Business Insights', 'Metrics & Reporting',
              'Experimentation', 'Inference', 'Stats & ML Modeling',
              'Model Deployment', 'ML Ops']

# Number of variables
num_vars = len(categories)

# Data for each role (adjust these values as needed)
data_engineers = [8, 9, 9, 4, 5, 4, 6, 3, 4, 5, 7, 8]
data_analysts = [3, 5, 6, 9, 9, 8, 7, 6, 5, 4, 3, 2]
data_scientists = [7, 7, 6, 9, 7, 9, 7, 7, 8, 9, 7, 9]
ml_engineers = [7, 6, 7, 4, 5, 4, 5, 4, 6, 8, 9, 9]

# Group the data
values = [
    data_engineers,
    data_analysts,
    data_scientists,
    ml_engineers
]

# Role labels
role_labels = ['Data Engineers', 'Data Analysts', 'Data Scientists', 'ML Engineers']

# Define colors
colors = ['green', 'red', 'blue', 'orange']

# Set up the radar chart
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Plot setup
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
plt.xticks(angles[:-1], categories, color='grey', size=8)
ax.set_rlabel_position(0)

# Plot each group with reduced fill colors and add data points on borders
for i, (role, color) in enumerate(zip(values, colors)):
    role += role[:1]
    ax.plot(angles, role, color=color, linewidth=2, linestyle='solid', label=role_labels[i])
    ax.fill(angles, role, color=color, alpha=0.15)  # Reduced alpha for less intensity
    ax.scatter(angles, role, color=color, s=50)  # Add data points on the borders


plt.xticks(angles[:-1], categories, color='grey', size=8, fontweight='bold')

# Add a legend
legend_elements = [Patch(facecolor=color, edgecolor=color, label=label) for color, label in zip(colors, role_labels)]
ax.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Display the plot
plt.show()
```
